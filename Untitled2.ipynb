```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.mixture import GaussianMixture as GMM
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
  
  
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
  
```

```python




df=pd.read_csv("/content/drive/MyDrive/ds3_project/Stock Headlines.csv", encoding= 'unicode_escape')
```

```python



Noe=df.isnull().sum()
print(Noe)
df = df.dropna(axis=0)
Noe=df.isnull().sum()
print(Noe)

 
```

```python

train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
data1=test.iloc[:,2:27]
data1.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

```

```python


list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)
# Convertng headlines to lower case
for index in new_Index:
    data[index]=data[index].str.lower()
    
data.head(1)

headlines = []
for row in range(0,len(data.index)):
    
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))

```

```python
list2= [i for i in range(25)]
new_Index=[str(i) for i in list2]
data1.columns= new_Index
data1.head(5)
# Convertng headlines to lower case
for index in new_Index:
    data1[index]=data1[index].str.lower()


headlines1 = []
for row in range(0,len(data1.index)):
    
    headlines1.append(' '.join(str(x) for x in data1.iloc[row,0:25]))




```

```python
headlines
len(headlines)
```

```python
w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()      
X=[] 
#ps = PorterStemmer()  
for x in headlines:  
    #tokens=(word_tokenize(x))
    tokens=w_tokenizer.tokenize(x)
    lemma=[] 
    for token in tokens:
     lemma_word=lemmatizer.lemmatize(token)   
     #stemmed_word = ps.stem(token)
     lemma.append(lemma_word)
    X.append(lemma)
Y=[]    
for j in X:    
    Y.append(' '.join(j))
#print(Y[0])    
Y
```

```python
Y
len(Y)
```

```python
X1=[] 
#ps = PorterStemmer()  
for x in headlines1:  
    #tokens=(word_tokenize(x))
    tokens=w_tokenizer.tokenize(x)
    lemma=[] 
    for token in tokens:
     lemma_word=lemmatizer.lemmatize(token)   
     #stemmed_word = ps.stem(token)
     lemma.append(lemma_word)
    X1.append(lemma)
Y1=[]    
for j in X1:    
    Y1.append(' '.join(j))
#print(Y[0])    
Y1
```

```python




countvector=CountVectorizer(ngram_range=(2,2))
traindataset1=countvector.fit_transform(Y)
type(traindataset1)
```

```python
## implement BAG OF WORDS
tfidfvector=TfidfVectorizer(ngram_range=(2,2))
traindataset=tfidfvector.fit_transform(Y)

```

```python
naive=MultinomialNB()
naive.fit(traindataset,train['Label'])

test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = tfidfvector.transform(Y1)
predictions = naive.predict(test_dataset)

```

```python
model = XGBClassifier()
model.fit(traindataset,train['Label'])
predictions4 = model.predict(test_dataset)
predictions4
```

```python
classifier = LogisticRegression(random_state = 0)
classifier.fit(traindataset,train['Label'])
predictions2=classifier.predict(test_dataset)
predictions2
```

```python
# Using K-means
knn5 = KNeighborsClassifier(n_neighbors=5)
knn5.fit(traindataset,train['Label'])
y_pred5 = knn5.predict(test_dataset)

```
